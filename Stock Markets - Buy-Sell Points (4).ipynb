In this project I created a mock dataset to make an example to show how we use the analytical approach of technical analysis and the numerical approach of simple moving averages (SMA) to find potential buy-sell points for a stock. A simple moving average (SMA) is an arithmetic moving average calculated by adding recent prices and then dividing that figure by the number of time periods in the calculation average.[test](https://www.investopedia.com/terms/s/sma.asp)

#### 1. Numerical Approach

We will use the popular Python library 'pandas' for data manipulation and 'matplotlib' for visualization. The results should be taken with caution since the data is random and doesn't represent actual stock price movements. In practice, we would use real historical stock price data to analyze and make decisions.



```python
!pip install pandas matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

#Now, we'll create a mock dataset with 365 days of stock prices

np.random.seed(42)
date_range = pd.date_range(start='2022-01-01', periods=365, freq='D')
stock_prices = np.random.normal(loc=100, scale=5, size=len(date_range))

stock_data = pd.DataFrame({'Date': date_range, 'Close': stock_prices})
stock_data.set_index('Date', inplace=True)

#Now, let's calculate the simple moving averages for 20-day (short-term) and 50-day (long-term) periods

stock_data['SMA20'] = stock_data['Close'].rolling(window=20).mean()
stock_data['SMA50'] = stock_data['Close'].rolling(window=50).mean()
 
#When the short-term moving average (SMA20) crosses above the long-term moving average (SMA50), it is considered a buy signal.
#Conversely, when the SMA20 crosses below the SMA50, it is considered a sell signal

def get_buy_sell_signals(data):
    buy_signals = []
    sell_signals = []
    position = False

    for i in range(len(data)):
        if data['SMA20'][i] > data['SMA50'][i] and not position:
            buy_signals.append(data['Close'][i])
            sell_signals.append(None)
            position = True
        elif data['SMA20'][i] < data['SMA50'][i] and position:
            sell_signals.append(data['Close'][i])
            buy_signals.append(None)
            position = False
        else:
            buy_signals.append(None)
            sell_signals.append(None)
    
    return buy_signals, sell_signals

stock_data['Buy_Signal'], stock_data['Sell_Signal'] = get_buy_sell_signals(stock_data)



plt.figure(figsize=(12, 6))
plt.plot(stock_data['Close'], label='Stock Price', alpha=0.5)
plt.plot(stock_data['SMA20'], label='20-Day Moving Average', linestyle='--')
plt.plot(stock_data['SMA50'], label='50-Day Moving Average', linestyle='--')
plt.scatter(stock_data.index, stock_data['Buy_Signal'], label='Buy Signal', marker='^', color='g')
plt.scatter(stock_data.index, stock_data['Sell_Signal'], label='Sell Signal', marker='v', color='r')
plt.title('Mock Stock Price with Buy-Sell Signals')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show
```

The graph above, shows when the SMA20 (short-term moving average) crosses above the SMA50 (long-term moving average), it is considered a buy signal. This crossover suggests that the stock price has gained upward momentum in the short-term, indicating a potential trend reversal or the beginning of a bullish trend. Investors who follow this strategy may view this as a good time to enter the market and purchase the stock, expecting it to continue rising in value.

Conversely, when the SMA20 crosses below the SMA50, it is considered a sell signal. This scenario implies that the stock's short-term momentum has shifted downward, possibly indicating a trend reversal or the beginning of a bearish trend. Investors who follow this strategy may interpret this as a signal to exit their position, selling the stock to avoid potential losses due to the anticipated decline in value.

For the analytical approach, since the dataset above is randomly generated, we are not able to use the same method analytically in otherwords,it would be impractical to calculate moving averages and buy-sell points manually without coding. However, we can discuss the general concept of moving averages and how to identify buy-sell points in a simplified example with a small set of data.

Let's assume we have the following closing prices for 10 days:

Day 1: $100

Day 2: $102

Day 3: $101

Day 4: $103

Day 5: $105

Day 6: $104

Day 7: $106

Day 8: $107

Day 9: $109

Day 10: $110

We will use 3-day (short-term) and 5-day (long-term) simple moving averages (SMA) to determine buy-sell points.

SMA3_Day4 = $\frac{(Day2 + Day3 + Day4)}{3} = \frac{(102 + 101 + 103)}{ 3 }= 102$

SMA3_Day5 = $\frac{(Day3 + Day4 + Day5) }{ 3} = \frac {(101 + 103 + 105)} { 3} = 103$

SMA3_Day6 = $\frac {(Day4 + Day5 + Day6) }{3} = \frac {(103 + 105 + 104)}{ 3} = 104$

...

Similarly, for 5-day SMA:

SMA5_Day6 = $\frac{(Day2 + Day3 + Day4 + Day5 + Day6)}{ 5} = \frac{(102 + 101 + 103 + 105 + 104) }{5 }= 103$

SMA5_Day7 = $\frac{ (Day3 + Day4 + Day5 + Day6 + Day7) /}{5} = \frac{(101 + 103 + 105 + 104 + 106)}{ 5} = 103.8$

...

Now, let's compare the short-term (3-day) and long-term (5-day) moving averages:

Day 4: SMA3 = 102, SMA5 = N/A (We need 5 days of data to calculate the 5-day SMA)

Day 5: SMA3 = 103, SMA5 = N/A

Day 6: SMA3 = 104, SMA5 = 103


we used a 3-day (short-term) and a 5-day (long-term) simple moving average to determine buy-sell points. The concept behind this approach is that when the short-term moving average crosses above the long-term moving average, it can indicate a potential buying opportunity, signaling that the stock is in an upward trend. Conversely, when the short-term moving average crosses below the long-term moving average, it can signal a potential selling opportunity, as the stock may be entering a downward trend. On Day 6, the short-term moving average (SMA3) crosses above the long-term moving average (SMA5). This is a buy signal. We would continue to calculate the moving averages for the subsequent days and observe for any crossover. When the short-term moving average crosses below the long-term moving average, it's a sell signal.

It's important to note that this example is highly simplified, and real-world stock data would require more sophisticated analysis, such as using longer moving average periods (like 20-day and 50-day) and other technical indicators. Additionally, keep in mind that past performance does not guarantee future results, and the moving average crossover strategy may not always provide accurate buy-sell signals. In practice, the choice of moving average periods depends on the investment horizon and the specific stock or market being analyzed. Longer moving averages (e.g., 50-day, 100-day, or 200-day) are typically used for identifying long-term trends, while shorter moving averages (e.g., 5-day, 10-day, or 20-day) are more suited for short-term trading or analyzing stocks with higher volatility.
